```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://internshala.com/jobs/page-1/"
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.text,"lxml")
soup=soup.prettify
```

```python
Name=soup.find_all("h3",class_="heading_4_5 profile")
Job_name=[]
for i in Name:
  name=i.text.strip()
  Job_name.append(name)

Job_name
```

```python
Name1=soup.find_all("div",class_="heading_6 company_name")
Company_name=[]
for i in Name1:
  name=i.text.strip()
  Company_name.append(name)
Company_name
```

```python
Name2=soup.find_all("a",class_="location_link view_detail_button")
location=[]
for i in Name2:
  name=i.text
  location.append(name)
location
```

```python
Name3=soup.find_all("span",class_="mobile")
salary=[]
for i in Name3:
  name=i.text.strip()
  salary.append(name)
salary
```

```python
Name4=soup.find_all("div",class_="item_body mobile-text")
exp=[]
for i in Name4:
  name=i.text.strip()
  exp.append(name)
exp
```

```python
Name5=soup.find_all("div",class_="status status-small status-inactive")
Job_type=[]
for i in Name5:
  name=i.text.strip()
  Job_type.append(name)
Job_type
```

```python
data = []
for i in range(1,118):
  url=f"https://internshala.com/jobs/page-{i}/"
  r=requests.get(url)
  soup=BeautifulSoup(r.text,"lxml")

  job_name1= soup.find_all("h3", class_="heading_4_5 profile")
  company_name1=soup.find_all("div", class_="heading_6 company_name")
  location_name=soup.find_all("a",class_="location_link view_detail_button")
  Salary_Rs=soup.find_all("span",class_="mobile")
  Experience=soup.find_all("div",class_="item_body mobile-text")
  Job_status=soup.find_all("div",class_="status status-small status-inactive")
  for name, company, Location, Salary, Experience, job_sta in zip(job_name1, company_name1, location_name, Salary_Rs, Experience, Job_status):
      job_name = name.text.strip()
      company_name = company.text.strip()
      location_name=Location.text.strip()
      salary=Salary.text.strip()
      experience=Experience.text.strip()
      job_i=job_sta.text.strip()
      data.append({"Job Name": job_name, "Company Name": company_name, "Location":location_name, "Salary":salary, "Experience":experience, "Job_Status":job_i})

```

```python
df1=pd.DataFrame(data)
```

```python
df1
```

```python
df1['Salary'] = df1['Salary'].str.replace('/year', '')
df1['Experience'] = df1['Experience'].str.replace(r'years? experience', '', case=False)
df1.head(15)
```

```python
df1.to_excel("Internshala.xlsx", index=False)
```
